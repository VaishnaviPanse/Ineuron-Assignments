1. Create the below pattern using nested for loop in Python.

*
**
***
****
*****
****
***
**
*

```python
n=int(input("enter value:"))
for i in range(n):
    print('*'*(i+1))

for i in range(n-1):
    print('*'*(n-i-1))
```
