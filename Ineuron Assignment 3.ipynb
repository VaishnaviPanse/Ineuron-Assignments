1.1Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def myreduce(num):
    
    num_list=list(range(1,number+1))
    sum_of_elements=0
    
    for i in num_list:
        sum_of_elements+=i
        
    return num_list,sum_of_elements
 

number=int(input(" insert the number :"))

output=myreduce(number)


print("List of First n Natural numbers are:",output[0])
print("Sum of List elements are :",output[1])
```
